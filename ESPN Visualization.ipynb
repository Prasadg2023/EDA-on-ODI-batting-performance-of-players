# Steps
## 1) Univariate Analysis (Single Column)

* Distribution of one variable

* Count, histogram, boxplot

## Relationship between two columns

### Examples:

* Numeric vs Numeric → scatter plot

* Numeric vs Categorical → box plot

* Categorical vs Categorical → count plot

# 1) Univariate Analysis - Statistical 
## I) For Numerical Columns

```python
df.columns
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64'])
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print(col_name)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(numerical_cols)
```

## II) For Categorical Columns

```python
categorical_cols = df.select_dtypes(include=['object'])
def categorical_univariate_analysis(categorical_data): 
    for col_name in categorical_data:
        print(col_name)
        print(categorical_data[col_name].agg(['count', 'nunique', 'unique']))
        print()
```

```python
categorical_univariate_analysis(categorical_cols)
```

# Univariate Plots
## I) Numerical Columns

```python
for col in numerical_cols: 
    # Plots
    plt.figure(figsize=(12, 4))
    
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=20, color="skyblue")
    plt.title(f"{col} - Distribution")
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color="lightgreen")
    plt.title(f"{col} - Boxplot")
    plt.tight_layout()
    plt.show()
```

# Univariate Plots
## II) Categorical Columns

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    top_values = df[col].value_counts().head(5)
    sns.barplot(x=top_values.values,y=top_values.index, hue = top_values.index,legend=True, palette = "cool")
    plt.title(f"Top 5 {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to datetime (assuming only year is given)
df['Start_Year'] = pd.to_datetime(df['Start_Year'], format='%Y', errors='coerce')
df['End_Year'] = pd.to_datetime(df['End_Year'], format='%Y', errors='coerce')

# Calculate career length in years
df['Career_Length'] = df['End_Year'].dt.year - df['Start_Year'].dt.year

# Create a year range for each player's active years
all_years = []
for i, row in df.iterrows():
    if pd.notnull(row['Start_Year']) and pd.notnull(row['End_Year']):
        all_years.extend(range(row['Start_Year'].year, row['End_Year'].year + 1))

# Count number of active players each year
yearly_counts = pd.Series(all_years).value_counts().sort_index()

# Plot trend
plt.figure(figsize=(12,6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', color='b')
plt.title("Number of Active Players Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Players Active")
plt.grid(True)
plt.show()
```

# Multivariate Analysis

```python
plt.figure(figsize=(10,6))
sns.regplot(
    data=df,
    x='Matches',
    y='Total_Runs',
    scatter=False,       
    color='red',        
    line_kws={'linewidth':2}
)
sns.scatterplot(
    data=df,
    x='Matches',
    y='Total_Runs',
    hue='Country',        
    size='Batting_Average',
    alpha=0.7,
    sizes=(40, 400)
)
plt.title("Matches vs Total Runs (One Regression Line, Color=Country, Size=Batting Avg)", fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Observation:-
* Players with more matches generally score higher total runs, showing a strong positive relationship between career length and total performance.

* Batting averages vary widely – some players maintain high averages even with fewer matches, while others achieve consistency over long careers.

* Country-wise clustering is visible, with top cricketing nations dominating the higher runs and match counts compared to smaller cricket nations.

```python
# heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numerical_cols.corr(), annot=True, cmap='coolwarm',fmt='.2f')
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Bubble Plot 
top_df = df.head(10)
plt.figure(figsize=(12,7))
plt.scatter(top_df['Matches'], top_df['Total_Runs'],
            s=top_df['Strike_Rate']*2, alpha=0.6, edgecolors='black', c=top_df['Batting_Average'], cmap='plasma')
for i in range(top_df.shape[0]):
    plt.text(top_df['Matches'].iloc[i]+0.5, top_df['Total_Runs'].iloc[i],
             top_df['Player_Names'].iloc[i], fontsize=8)
plt.colorbar(label="Batting Average")
plt.xlabel("Matches")
plt.ylabel("Total Runs")
plt.title("Bubble Plot: Matches vs Runs (Color=Batting Avg, Size=Strike Rate)", fontsize=16)
plt.show()
```

# Observation:-
* More matches = more runs – Players who played more matches generally scored more runs.

* Kohli has best average – Even with fewer matches, Virat Kohli has the highest batting average.

* Big hitters have high strike rates – Players like Sehwag and Gayle score faster, shown by larger bubbles.
